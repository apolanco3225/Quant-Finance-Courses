```python
summary_df = returns_df.describe().T.loc[:, ["mean", "std"]]
summary_df.plot.scatter(x="std", y="mean", figsize=(15, 5), s=50, fontsize=15)
off_value = 0.000001
for company in summary_df.index:
    coordinates = (
    summary_df.loc[company, "std"] + off_value, 
    summary_df.loc[company, "mean"] + off_value
    )
    print(company, coordinates)
    plt.annotate(
        text=company, 
        xy = coordinates, 
        size=15,
    )

plt.title("Risk - Return")
plt.xlabel("Annual Risk")
plt.ylabel("Annual Return")
plt.show();
```

# Cumulative Returns, Drawdowns, etc. 

```python
apple_df = yf.download("AAPL")
```

```python
apple_df = apple_df.Close.to_frame()
apple_df.head()
```

```python
apple_df["daily_returns"] = np.log(apple_df.Close.div(apple_df.Close.shift(1)))
apple_df["daily_returns"].head()
```

```python
apple_df.dropna(inplace=True)
```

```python
apple_df.daily_returns.sum()
```

```python
np.exp(apple_df.daily_returns.sum())
```

```python
apple_df["cumulative_returns"] = apple_df.daily_returns.cumsum().apply(np.exp)
apple_df.head()
```

```python
apple_df.cumulative_returns.plot()
```

```python
apple_df.daily_returns.mean() * 255
```

```python
apple_df.daily_returns.std() * np.sqrt(255)
```
