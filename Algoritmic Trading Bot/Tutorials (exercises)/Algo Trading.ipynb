```python
apple_df = close_df["AAPL"].copy().to_frame()
apple_df.head()
```

```python
apple_df.shift().head()
```

```python
apple_df.shift(-1).head()
```

```python
apple_df["lag_1"] = apple_df["AAPL"].shift(periods=1)
apple_df.head()
```

```python
apple_df["diff"] = apple_df["AAPL"].sub(apple_df["lag_1"])
apple_df.head()
```

```python
apple_df.columns = [column.lower() for column in apple_df.columns]
apple_df.head()
```

```python
apple_df["percentage_change"] = apple_df.aapl.div(apple_df.lag_1)
apple_df["diff_2"] = apple_df.aapl.diff(periods=1)
apple_df["percentage_change_2"] = apple_df.aapl.pct_change(periods=1).mul(100)
apple_df.head()
```

```python
apple_df = apple_df[["aapl", "percentage_change_2"]]
apple_df.head()
```

```python
ticker_list = ["SPY", "AAPL", "KO", "IBM", "DIS", "MSFT"]

stocks_df = yf.download(
    ticker_list, 
    start="2010-01-01", 
    end="2021-01-01"
)

stocks_df.head()
```

```python
close_df = stocks_df["Close"]
close_df.plot(figsize=(12, 8));
```

```python
returns_df = close_df.pct_change().fillna(0)
returns_df.plot(figsize=(12, 8));
```

```python
returns_df.head()
```

```python
returns_df.describe().T
```

```python
returns_df.describe().T.loc[:, ["mean", "std"]]
```

```python
summary_df = returns_df.describe().T.loc[:, ["mean", "std"]]
summary_df.plot.scatter(x="std", y="mean", figsize=(15, 5), s=50, fontsize=15)
off_value = 0.000001
for company in summary_df.index:
    coordinates = (
    summary_df.loc[company, "std"] + off_value, 
    summary_df.loc[company, "mean"] + off_value
    )
    print(company, coordinates)
    plt.annotate(
        text=company, 
        xy = coordinates, 
        size=15,
    )

plt.title("Risk - Return")
plt.xlabel("Annual Risk")
plt.ylabel("Annual Return")
plt.show();
```
