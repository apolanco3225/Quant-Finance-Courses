```python
data["Date"] = pd.to_datetime(data["Date"])
data.info()
```

```python
data = pd.read_csv(
    filepath_or_buffer="data/walmart_stock.csv",
    parse_dates=["Date"], 
    index_col="Date"
)
data.head()
```

```python
data.info()
```

```python
data.resample(rule="A").mean()
```

```python
data.resample(rule="Q").mean().head()
```

```python
data.resample(rule="BQ").mean().head()
```

```python
data.resample(rule="A").mean().max()
```

```python
def first_day(entry):
    return entry[0]

data.resample(rule="A").apply(first_day)
```
